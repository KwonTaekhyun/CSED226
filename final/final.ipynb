## Extra Tree Classifier

```python
from sklearn.ensemble import BaggingClassifier
from sklearn.tree import ExtraTreeClassifier
from sklearn.ensemble import RandomForestClassifier
```

Random forest

```python
X_train = train_df[['Elevation', 'Aspect', 'Slope', 'Hillshade_Morning',
                    'Hillshade_Afternoon', 'Hillshade_Evening', 'Distance_To_River',
                    'Distance_From_Road', 'Distance_From_Pollutant', 'Sector', 'Soil']]
y_train = train_df[['Plant_species']]

X_test = test_df

random_forest = RandomForestClassifier(
    n_estimators=255, criterion='entropy', warm_start=True)
random_forest.fit(X_train, y_train)

y_pred = random_forest.predict(X_test)

y_pred_df = pd.DataFrame()
y_pred_df['Id'] = np.arange(20000)
y_pred_df['Category'] = y_pred

y_pred_df.to_csv('./submission/CLF_RF_E255_CE_WS.csv', index=False)
```

Extra tree classifier

```python
X_train = train_df[['Elevation', 'Aspect', 'Slope', 'Hillshade_Morning',
                    'Hillshade_Afternoon', 'Hillshade_Evening', 'Distance_To_River',
                    'Distance_From_Road', 'Distance_From_Pollutant', 'Sector', 'Soil']]
y_train = train_df[['Plant_species']]

X_test = test_df

extra_tree = ExtraTreeClassifier(random_state=0)

extra_tree.fit(X_train, y_train)

y_pred = extra_tree.predict(X_test)

y_pred_df = pd.DataFrame()
y_pred_df['Id'] = np.arange(20000)
y_pred_df['Category'] = y_pred

y_pred_df.to_csv('./submission/ET.csv', index=False)
```

Extra tree classifier with Bagging

```python
X_train = train_df[['Elevation', 'Aspect', 'Slope', 'Hillshade_Morning',
                    'Hillshade_Afternoon', 'Hillshade_Evening', 'Distance_To_River',
                    'Distance_From_Road', 'Distance_From_Pollutant', 'Sector', 'Soil']]
y_train = train_df[['Plant_species']]

X_test = test_df

extra_tree = ExtraTreeClassifier(random_state=0, criterion="entropy", splitter="best", class_weight="balanced")

model = BaggingClassifier(extra_tree, random_state=0,n_estimators=200).fit(
    X_train, y_train)

y_pred = model.predict(X_test)

y_pred_df = pd.DataFrame()
y_pred_df['Id'] = np.arange(20000)
y_pred_df['Category'] = y_pred

y_pred_df.to_csv('./submission/ET_B_Ce_Sb_CWb_NE200.csv', index=False)
```
